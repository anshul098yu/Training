```python
from pyspark.sql.types import *
from pyspark.sql.functions import *
from pyspark.sql.window import *
```

```python
schema = StructType([
    StructField("order_id", StringType(), True),
    StructField("order_date", DateType(), True),
    StructField("customer_id", StringType(), True),
    StructField("product_id", StringType(), True),
    StructField("category", StringType(), True),
    StructField("region", StringType(), True),
    StructField("amount", DoubleType(), True)
])
```

```python
df=spark.read.format("csv").schema(schema).option('header',True).option("sep",",").load('/Volumes/dbw_de_training/anshul_m10/pipeline/read/online_sales_4.csv')
```

```python
df4=df.write.format('parquet').outputMode('append').option("path","/Volumes/dbw_de_training/anshul_m10/pipeline/Bronze/").option("checkpointLocation",
"/Volumes/dbw_de_training/anshul_m10/pipeline/Checkpoint/").start()
```

```python
df_bronze = spark.read.format("parquet").schema(schema).load("/Volumes/dbw_de_training/anshul_m10/pipeline/Bronze/")

from pyspark.sql.functions import current_timestamp
df_bronze = df_bronze.withColumn("currtimestamp", current_timestamp())

display(df_bronze)
```
